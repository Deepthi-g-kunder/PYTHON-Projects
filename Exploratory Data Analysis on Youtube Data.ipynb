### Influence of Time and Day of a week over views

```python
df['hour']=df.published_at.dt.hour
df['week'] = df.published_at.dt.day_name()
df['week_no'] = df.published_at.dt.weekday
```

```python
plt.figure(figsize=(100,5))
ax= pd.pivot_table(data=df,index='hour',aggfunc={'view_count':sum}).plot(kind='line')
ax.set_xticks(range(24))
plt.show()
```

View rate fluctuate a lot throghout the day with relatively high views from 4AM to 5AM  and between 3PM to 8PM and low view rate is observed in the first half of a day at multiple hours compared to second half ( post 12 PM)

```python
plt.figure(figsize=(100,5))
ax= pd.pivot_table(data=df,index=['week','week_no'],aggfunc={'view_count':sum}).sort_values(by='week_no').plot(kind='bar')
```

People watch more Youtube content during Friday than all other days

### Day and Hour wise Views Fluctuation

```python
pd.pivot_table(data=df,index =['week'],columns='hour',values='view_count',aggfunc=sum)
```

```python
# plt.figure(figsize=(15,15))

l= df.week.unique()

for x in l:
    d= df[df['week']==x]
    a=pd.pivot_table(data=d, index=['hour'],values='view_count',aggfunc=sum).reset_index()
    ax =sns.lineplot(data =df , x=a.hour,y=a.view_count,color ='red')
    ax.set_xticks(range(24))
    ax.set_title(x)
    plt.show()
```

More view is observed in the second half of day in all the days of a week except Friday, Where heavy spike observed during early morning and surprisingly low view rate during second half of the day.
